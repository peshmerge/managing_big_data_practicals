```python
import matplotlib
import matplotlib.pyplot as plt
# x axis values
x = [4,8,12,16,20]
# corresponding y axis values
y = [467,331,263,205,205]
 
matplotlib.rc('figure', figsize=(10, 6))

# plotting the points
plt.plot(x, y, 'ro--')
 
# naming the x axis
plt.xlabel('no. of executors')
# naming the y axis
plt.ylabel('time (in sec)')
 
for i, j in zip(x, y):
    plt.text(i+0.25, j+1.5, '(exec={}, {}s)'.format(i, j))

# giving a title to my graph
plt.title('No. of executer vs time taken for same data')
 
# function to show the plot
plt.show()
```
